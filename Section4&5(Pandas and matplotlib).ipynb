```python
import pandas as pd
```

```python
df = pd.read_csv("dataSet.csv")
```

```python
# Filtering
filtered = df[df["Score"] > 80]
```

```python
filtered
```

```python
# Grouping
grouped = df.groupby("Department")["Score"].mean()
```

```python
grouped
```

```python
# Sorting
sorted_df = df.sort_values(by="Score", ascending=False)
```
